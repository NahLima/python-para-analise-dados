```python
#Considere o seguinte código:
import pandas as pd
data = [['Fulano', 12, 7.0, True],
        ['Sicrano', 15, 3.5, False], 
        ['Beltrano', 18, 9.3, True]]
dados2 = pd.DataFrame(data, 
        columns = ['Aluno', 'Idade', 'Nota', 'Aprovado'])
dados2
```

```python
# Para obtermos uma tabela contendo os nomes das variáveis 
# e seus respectivos tipos de dados, conforme o exemplo abaixo, 
# que linhas de código devemos executar no Jupyter?

tipos_de_dados= pd.DataFrame(dados2.dtypes, columns=['Tipos de Dados'])
tipos_de_dados.columns.name = 'Variaveis'
tipos_de_dados
```

### Como importar a biblioteca: 
 import pandas as pd

### Como ler fontes de dados diferentes:
- Uma base CSV (pd.read_csv(...))

- Uma página HTML (pd.read_html(...))
