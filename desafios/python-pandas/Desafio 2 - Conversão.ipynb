### Desafio: Conversão

---

**Definição do problema:** Dada a tabela com a média de altura e peso de mulheres estadunidenses, onde os nomes das colunas estão em inglês *height* (altura) e *weight* (peso) e seus respectivos valores estão nas seguintes unidades: polegada e libras. 

<ol>
    <li> Construa uma nova representação onde os nomes das colunas estão em português, e os valores de altura são convertidos para centímetros e os valores de peso convertidos para quilograma. </li>
    <li> Plote o scatterplot de altura x peso da nova representação. </li>
</ol>

**Código do dataset:** women

**Dicas:**
<ul>
    <li> 1 polegada = 2,54 centímetros </li>
    <li> 1 libra = 0,45 quilo </li>
    <li> use o método .plot.scatter() </li>
</ul>

---

**Passo 1:** Importar as bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data 
```

**Passo 2:** Importar o dataset e verificar sua estrutura 

```python
''' comando que descreve o dataset em questão ''' 
data('women', show_doc=True)
```

```python
''' copia o dataset e verifica sua estrutura ''' 
df = data('women')
df
```

**Passo 3:** Construa uma nova representação onde os nomes das colunas estão em português, e os valores de altura são convertidos para centímetros e os valores de peso convertidos para quilograma.

```python
''' Solução 1: Como a conversão é simples, e no pandas ao multiplicar uma coluna (Series) por uma constante, 
todo o item da coluna é multiplicado pela constante, podemos fazer da seguinte forma: 
'''
df_conv1 = pd.DataFrame(columns = ['Altura', 'Peso'])
df_conv1['Altura'] = df['height']*2.54
df_conv1['Peso'] = df['weight']*0.45
df_conv1
```

```python
''' Solução 2: Nesta solução, ideal quando o processamento requerido é mais complexo, além de deixar o código 
mais organizado, pois para cada conversão exigida é criada uma função. '''

# Função que converte polegadas para centímetros
def inch_to_cm(x):
    return x*2.54

# Função que converte libras para quilogramas
def lbs_to_kg(x):
    return x*0.45

df_conv2 = pd.DataFrame(columns = ['Altura', 'Peso'])
df_conv2['Altura'] = df['height'].apply(inch_to_cm)
df_conv2['Peso'] = df['weight'].apply(lbs_to_kg)
df_conv2
```

**Passo 4:** Plote o scatterplot de altura x peso da nova representação.

Para compreender melhor a função de um scatterplot acesse: <ref> https://pt.wikipedia.org/wiki/Gr%C3%A1fico_de_dispers%C3%A3o </ref>

Utilizando o scatterplot é possível, de maneira visual, verificar se há correlação (neste caso relação linear) entre duas váriaveis.

```python
df_conv1.plot.scatter(x="Altura", y="Peso")
plt.show()
```

```python
df.plot.scatter(x="height", y="weight")
plt.show()
```
