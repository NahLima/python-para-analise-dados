```python
import pandas as pd
import seaborn as sns
#!pip install pydataset
from pydataset import data
```

# Boxplot - Exemplo 1

```python
''' Salva o DataFrame na variável e imprime as 10 primeiras ocorrências '''
df_mammals = data('Mammals')
df_mammals.head(10)
```

```python
# Agrupa as entradas pela variável hopper e gera um sumário da base.
df_mammals.groupby('hoppers')['speed'].describe()
```

```python
# Gera o boxplot para essas duas categorias.
sns.boxplot(data=df_mammals, x='hoppers', y='speed')
```

# Base complexa

```python
''' Salva o DataFrame na variável e imprime as 10 primeiras ocorrências '''
df = data('Forbes2000')
df.head(10)
```

```python
# Filtra a base de dados para as três categorias.
df_filtered = df[df['category'].isin(['Banking', 'Conglomerates', 'Insurance'])]

# Agrupa as entradas pela variável categoria e gera um sumário da base.
df_filtered.groupby('category')['marketvalue'].describe()
```

```python
# Gera o boxplot para essas três categorias.
sns.boxplot(data=df_filtered, x='category', y='marketvalue')
```
